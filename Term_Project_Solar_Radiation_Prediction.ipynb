```python
# Import the necessary modules
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import FunctionTransformer

# Define a function that applies re.search on a column and returns a new column
def regex_transformer(column, pattern):
    # column is a pandas Series object
    # pattern is a regular expression string
    # return a new pandas Series object with the matched strings
    return column.apply(lambda x: re.search(pattern, x).group(0) if re.search(pattern, x) else None)

# Define the column transformer with the desired transformations
ct = ColumnTransformer([
    # Extract the month from the Data column and drop the original column
    ("month", FunctionTransformer(regex_transformer, kw_args={"pattern": r'^\d+'}), "Data"),
    # Extract the day name from the Data column and drop the original column
    ("day", FunctionTransformer(regex_transformer, kw_args={"pattern": r'(?<=\/)\d+(?=\/)'}), "Data"),
    # Extract the hours from the Time column and drop the original column
    ("Time_Hr", FunctionTransformer(regex_transformer, kw_args={"pattern": r'^\d+'}), "Time"),
    # Extract the hours from the Time column and drop the original column
    ("Time_Min", FunctionTransformer(regex_transformer, kw_args={"pattern": r'(?<=:)\d+(?=:)'}), "Time"),
    # Extract the hours from the Time column and drop the original column
    ("Time_Sec", FunctionTransformer(regex_transformer, kw_args={"pattern": r'\d+$'}), "Time")
], remainder="passthrough")

# Create a pipeline with the column transformer
pipeline = Pipeline([
    ("ct", ct)
])

# Apply the pipeline on the dataframe and get the transformed dataframe
df_transformed = pipeline.fit_transform(data)
print(df_transformed)
```
